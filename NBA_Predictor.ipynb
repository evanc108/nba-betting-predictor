```python
import pandas as pd
import sqlite3
```

```python
bet_db = 'basketball-final.sqlite'
conn = sqlite3.connect(bet_db)
```

```python
game_stats = 'SELECT * FROM Game_FullDetails'
game_stats_df = pd.read_sql_query(game_stats, conn)
print(game_stats_df.head())
game_stats_df.dtypes
```

```python
game_stats_df["WL_AWAY"] = game_stats_df["WL_AWAY"].astype("category")
game_stats_df["Away_Outcome"] = game_stats_df["WL_AWAY"].cat.codes

game_stats_df["WL_HOME"] = game_stats_df["WL_HOME"].astype("category")
game_stats_df["Home_Outcome"] = game_stats_df["WL_HOME"].cat.codes
```

```python
game_stats_df.head()
```

```python
game_stats_df["Target"] = game_stats_df["PTS_HOME_y"]
```

```python
import numpy as np
```

```python
game_stats_df['Home/Away'] = game_stats_df.apply(lambda row: np.array([row['TEAM_ID_HOME'], row['TEAM_ID_AWAY']]), axis=1)
```

```python
game_stats_df["Home/Away"]
```

```python
game_stats_df["Away Opponent"] = pd.to_numeric(game_stats_df["TEAM_ID_AWAY"]).astype(int)

game_stats_df["Home Team"] = pd.to_numeric(game_stats_df["TEAM_ID_HOME"]).astype(int)

```

```python
game_stats_df.head()
```

```python
game_stats_df['GAME_DATE'] = pd.to_datetime(game_stats_df['GAME_DATE'])
```

```python
def previous_25_avg(group, field):
    group['Averaged ' + field] = group[field].shift().rolling(window=15, min_periods=1).mean()
    return group

fields_to_avg = [
    'PTS_HOME_x', 'PTS_AWAY', 'REB_HOME', 'REB_AWAY', 'AST_HOME', 'AST_AWAY',
    'STL_HOME', 'STL_AWAY', 'BLK_HOME', 'BLK_AWAY', 'TOV_HOME', 'TOV_AWAY',
    'FGM_AWAY', 'FGA_AWAY', 'FG_PCT_AWAY', 'FG3M_AWAY', 'FG3A_AWAY',
    'FG3_PCT_AWAY', 'FTM_AWAY', 'FTA_AWAY', 'FT_PCT_AWAY', 'OREB_AWAY',
    'DREB_AWAY', 'PF_AWAY',
    'FGM_HOME', 'FGA_HOME', 'FG_PCT_HOME', 'FG3M_HOME', 'FG3A_HOME',
    'FG3_PCT_HOME', 'FTM_HOME', 'FTA_HOME', 'FT_PCT_HOME', 'OREB_HOME',
    'DREB_HOME', 'PF_HOME'
]

# Calculate rolling averages for each field
for field in fields_to_avg:
    game_stats_df = game_stats_df.groupby('Team', group_keys=False).apply(lambda x: previous_25_avg(x, field))

# Drop rows with any NaN values
game_stats_df.dropna(how='any', inplace=True, subset=['Averaged ' + field for field in fields_to_avg])
```

```python
game_stats_df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = game_stats_df[game_stats_df["GAME_DATE"] < '2017-01-01']
```

```python
test = game_stats_df[game_stats_df["GAME_DATE"] > '2017-01-01']
```

```python
predictors = [
    "Averaged PTS_HOME_x", "Averaged PTS_AWAY", 
    "Averaged REB_HOME", "Averaged REB_AWAY", 
    "Averaged TOV_AWAY", "Averaged TOV_HOME", 
    "Averaged STL_AWAY", "Averaged STL_HOME", 
    "Averaged BLK_HOME", "Averaged BLK_AWAY",
    "Averaged FGM_AWAY", "Averaged FGA_AWAY", "Averaged FG_PCT_AWAY", 
    "Averaged FG3M_AWAY", "Averaged FG3A_AWAY", "Averaged FG3_PCT_AWAY", 
    "Averaged FTM_AWAY", "Averaged FTA_AWAY", "Averaged FT_PCT_AWAY", 
    "Averaged OREB_AWAY", "Averaged DREB_AWAY", "Averaged PF_AWAY",
    "Averaged FGM_HOME", "Averaged FGA_HOME", "Averaged FG_PCT_HOME", 
    "Averaged FG3M_HOME", "Averaged FG3A_HOME", "Averaged FG3_PCT_HOME", 
    "Averaged FTM_HOME", "Averaged FTA_HOME", "Averaged FT_PCT_HOME", 
    "Averaged OREB_HOME", "Averaged DREB_HOME", "Averaged PF_HOME"
]

```

```python
rf.fit(train[predictors], train["Target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
def custom_accuracy(y_true, y_pred, threshold=5):
    """
    Calculate custom accuracy where predictions are considered successful
    if they are within 'threshold' points of the actual targets.
    """
    success = abs(y_true - y_pred) <= threshold
    return success.mean()

# Assuming 'test' is your DataFrame and 'preds' are your model's predictions
# 'Target' is the actual target values

acc = custom_accuracy(test["Target"], preds)
```
