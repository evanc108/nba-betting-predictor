```python
import pandas as pd
```

```python
matches = pd.read_csv("https://raw.githubusercontent.com/evanc108/nba-betting-predictor/main/NBA%20Game%20Data/games.csv", header=0)
```

```python
matches.head()
```

```python
new_data = pd.DataFrame(columns=[
    'Date', 'Game_ID', 'Game_Status',
    'Team', 'Opponent', 'Location',
    'Season', 'PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST', 'REB',
    'Outcome'
])
```

```python
for index, row in matches.iterrows():
    home_team_data = [
        row['GAME_DATE_EST'], row['GAME_ID'], row['GAME_STATUS_TEXT'],
        row['HOME_TEAM_ID'], row['VISITOR_TEAM_ID'],
        'Home', row['SEASON'], row['PTS_home'], row['FG_PCT_home'], row['FT_PCT_home'],
        row['FG3_PCT_home'], row['AST_home'], row['REB_home'],
        'Won' if row['HOME_TEAM_WINS'] == 1 else 'Lost'  # Winner determination
    ]
    
    away_team_data = [
        row['GAME_DATE_EST'], row['GAME_ID'], row['GAME_STATUS_TEXT'],
        row['VISITOR_TEAM_ID'], row['HOME_TEAM_ID'],
        'Away', row['SEASON'], row['PTS_away'], row['FG_PCT_away'], row['FT_PCT_away'],
        row['FG3_PCT_away'], row['AST_away'], row['REB_away'],
        'Lost' if row['HOME_TEAM_WINS'] == 1 else 'Won'  # Winner determination
    ]
    
    home_team_df = pd.DataFrame([home_team_data], columns=new_data.columns)
    away_team_df = pd.DataFrame([away_team_data], columns=new_data.columns)
    
    new_data = pd.concat([new_data, home_team_df, away_team_df], ignore_index=True)

new_data.to_csv('reorganized_data.csv', index=False)
```
