```python
import pandas as pd
```

```python
matches = pd.read_csv("https://raw.githubusercontent.com/evanc108/nba-betting-predictor/main/NBA%20Game%20Data/games.csv", header=0)
```

```python
matches.head()
```

```python
new_data = pd.DataFrame(columns=[
    'Date', 'Game_ID', 'Game_Status',
    'Team', 'Opponent', 'Location',
    'Season', 'PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST', 'REB',
    'Outcome'
])
```

```python
for index, row in matches.iterrows():
    home_team_data = [
        row['GAME_DATE_EST'], row['GAME_ID'], row['GAME_STATUS_TEXT'],
        row['HOME_TEAM_ID'], row['VISITOR_TEAM_ID'],
        'Home', row['SEASON'], row['PTS_home'], row['FG_PCT_home'], row['FT_PCT_home'],
        row['FG3_PCT_home'], row['AST_home'], row['REB_home'],
        'Won' if row['HOME_TEAM_WINS'] == 1 else 'Lost'  # Winner determination
    ]
    
    away_team_data = [
        row['GAME_DATE_EST'], row['GAME_ID'], row['GAME_STATUS_TEXT'],
        row['VISITOR_TEAM_ID'], row['HOME_TEAM_ID'],
        'Away', row['SEASON'], row['PTS_away'], row['FG_PCT_away'], row['FT_PCT_away'],
        row['FG3_PCT_away'], row['AST_away'], row['REB_away'],
        'Lost' if row['HOME_TEAM_WINS'] == 1 else 'Won'  # Winner determination
    ]
    
    home_team_df = pd.DataFrame([home_team_data], columns=new_data.columns)
    away_team_df = pd.DataFrame([away_team_data], columns=new_data.columns)
    
    new_data = pd.concat([new_data, home_team_df, away_team_df], ignore_index=True)

new_data.to_csv('reorganized_data.csv', index=False)
```

```python
new_data.head(20)
```

```python
nba_df = new_data;
```

```python
nba_df.head()
```

```python
nba_df["Target"] = (nba_df["Outcome"] == "Won").astype("int")
```

```python
nba_df["location_code"] = nba_df["Location"].astype("category").cat.codes
```

```python
nba_df["opp_code"] = nba_df["Opponent"].astype("category").cat
```

```python
nba_df.dtypes
```

```python
nba_df["Opponent"] = pd.to_numeric(nba_df["Opponent"]).astype(int)
```

```python
nba_df["Team"] = pd.to_numeric(nba_df["Team"]).astype(int)
```

```python
nba_df.dtypes
```

```python
nba_df.head()
```

```python
nba_df['Date'] = pd.to_datetime(nba_df['Date'])
```

```python
def previous_25_avg(group, field):
    group['Averaged ' + field] = group[field].shift().rolling(window=40, min_periods=1).mean()
    return group

fields_to_avg = ['PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST', 'REB']

# Calculate rolling averages for each group
nba_df = nba_df.groupby('Team', group_keys=False).apply(lambda x: previous_25_avg(x, field))

# Drop rows with any NaN values
nba_df.dropna(how='any', inplace=True, subset=['Averaged ' + field for field in fields_to_avg])
```

```python
nba_df.head()
```

```python
nba_df.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = nba_df[nba_df["Date"] < '2021-01-01']
```

```python
test = nba_df[nba_df["Date"] > '2021-01-01']
```

```python
predictors = ["location_code", "Opponent", "Averaged PTS", "Averaged FG_PCT", "Averaged FT_PCT", "Averaged FG3_PCT", "Averaged AST", "Averaged REB"]
```

```python
rf.fit(train[predictors], train["Target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["Target"], preds)
```
